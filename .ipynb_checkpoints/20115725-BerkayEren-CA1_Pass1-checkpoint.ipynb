This notebook applies the three-pass Exploratory Data Analysis (EDA) approach to a dataset describing released prisoners and whether they re-offended within 12 months.

```python
# =============================================================
# PASS 1: LOADING DATA & INITIAL CLEANING 
# =============================================================

# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

sns.set(style="whitegrid")
```

### Import Libraries
- **pandas, numpy** → Data handling and numeric operations  
- **matplotlib, seaborn** → Data visualisation  
- **missingno** → Visualising missing data  
- **sns.set()** ensures consistent plot styling


```python
# read dataset with tab separator
df = pd.read_csv("data/arrests.csv", sep="\t")
df.head()
```

### About the Dataset
Each row represents one individual released from prison.  
The **target variable** `tgt` indicates whether the person **reoffended** (`y`) or **did not reoffend** (`n`).  

Feature names such as `a1–a4`, `c1–c3`, and `d1–d5` are anonymised but grouped to represent related aspects such as demographics, behaviour, or background characteristics.


```python
df.info()
```

### Dataset Structure
- **60,458** observations and **21** columns  
- 3 numeric features (`a1`, `d1`, `d2`)  
- 18 categorical features  
- Several columns contain missing values (`a3`, `c1`, `c2`, `e1`, `e2`, `g`)  
This gives an early indication of which features may need cleaning.


```python
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Values': missing,
    'Percentage (%)': missing_pct.round(2)
})

missing_summary
```

```python
plt.figure(figsize=(12,6))
msno.matrix(df)
plt.title("Missing Data Overview")
plt.show()
```

### Missing Data Overview
- **High missingness (>80%)** in `e1`, `e2`, and `g`.  
- **Moderate missingness (15–40%)** in `a3`, `c1`, and `c2`.  
- **Numeric columns** (`a1`, `d1`, `d2`) have no missing values.  

The missingness appears random rather than systematic, meaning it likely won’t bias results after imputation.
