The aim of Pass 3 is to **investigate how features relate to one another and to the target variable (`tgt`)**.

The objectives are to:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import combinations
from scipy.stats import chi2_contingency

sns.set(style="whitegrid")

# reload cleaned dataset
df = pd.read_csv("data/cleaned_arrests.csv")

# re-convert categorical types
cat_cols = df.select_dtypes(include='object').columns.tolist()
df[cat_cols] = df[cat_cols].astype('category')

df.info()
```

### Correlation Heatmap Among Numeric Features

```python
num_cols = ['a1', 'd1', 'd2']

plt.figure(figsize=(5,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='Reds', fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Numeric Features")
plt.show()
```

The heatmap shows how the numeric features relate to each other.  
`a1`and `d2` have a **moderately strong correlation (around 0.69)**,  
meaning they often increase or decrease together and may describe similar behaviour.  
`d1` is only **weakly related** to the others,  
so it adds some new information to the dataset.  

Overall, the numeric features are not fully independent but still distinct enough  
to be useful together in a predictive model.

### Relationships Within Feature Groups
Examining feature groups (A, C, D) helps to see whether features inside the same group  
describe similar information or independent aspects.

```python
encoded_df = df.copy()
for col in encoded_df.select_dtypes(include='category').columns:
    encoded_df[col] = encoded_df[col].cat.codes

groups = {
    "A-group": ['a1', 'a2', 'a3', 'a4'],
    "C-group": ['c1', 'c2', 'c3'],
    "D-group": ['d1', 'd2', 'd3', 'd4', 'd5']
}

for name, cols in groups.items():
    plt.figure(figsize=(6,4))
    sns.heatmap(encoded_df[cols].corr(), annot=True, cmap='Reds', fmt=".2f", linewidths=0.5)
    plt.title(f"Correlation Within {name}")
    plt.show()
```

**A-group (`a1â€“a4`)**  
- Very weak relationships between all variables (all below 0.15).  
- Each feature seems to describe a **separate aspect** of behaviour.  
- No risk of duplication â€” all can be safely included in later analysis.

**C-group (`c1â€“c3`)**  
- Mostly weak relationships.  
- A small **negative link between `c2` and `c3` (-0.23)** suggests that when one increases, the other slightly decreases.  
- Overall, the group remains **fairly independent**.

**D-group (`d1â€“d5`)**  
- Some moderate internal connections:  
  - `d1` and `d2` show a **positive correlation of 0.44**,  
  - `d1` and `d3` about **0.36**.  
- These patterns indicate that **`d1` may represent a broader factor** shared by several `d` variables.  
- `d4` and `d5` stay almost unrelated to others, acting as more **distinct features**.

**Summary**  
Only the D group shows moderate overlap, while A and C groups are largely independent.  
This means most features likely provide **unique information** useful for modelling,  
though `d1â€“d3` could carry **partly redundant content**.

### Relationships Among Categorical Features

To understand how categorical features relate to each other,  
cross-tab plots can reveal whether certain categories often appear together.  
The example below compares category combinations between selected features.

```python
pairs = [
    ('a2', 'a3'),
    ('a3', 'a4'),
    ('c1', 'c2'),
    ('c2', 'c3'),
    ('f1', 'f2')
]

plt.figure(figsize=(14,10))
for i, (x, y) in enumerate(pairs, 1):
    plt.subplot(3, 3, i)
    cross_tab = pd.crosstab(df[x], df[y])
    sns.heatmap(cross_tab, cmap='Reds')
    plt.title(f"{x} vs {y}")
    plt.xlabel(y)
    plt.ylabel(x)
plt.tight_layout()
plt.show()
```

The heatmaps show how category combinations occur across different categorical features.

**A-group (`a2â€“a4`)**
- `a2` and `a3` display several dense horizontal bands, meaning some categories co-occur often.  
  â†’ Moderate internal relationship, some overlap of information.
- `a3` and `a4` are dominated by one `a4` category, visible as a vertical stripe.  
  â†’ Indicates `a4` is less diverse and less dependent on `a3`.

**C-group (`c1â€“c3`)**
- `c1` and `c2` show moderate clustering but no single dominant diagonal.  
  â†’ Some shared structure but still distinct.  
- `c2` and `c3` appear mostly independent, with weak scattered colour intensity.  
  â†’ Likely capture different characteristics.

**F-group (`f1â€“f2`)**
- A single dark square dominates the plot.  
  â†’ These two features are **strongly related or partially redundant**, describing almost the same condition.

**Overall**
- A and C groups show mild internal associations,  
  while F group variables are almost identical.  
- Most categorical features still vary independently enough to add new information to the model.


### Relationships Between Features and Target (`tgt`)
This section examines how each feature differs between individuals who **reoffended (`y`)**
and those who **did not (`n`)**.
Patterns of imbalance or separation help identify which variables are likely to
be **predictive** in a future model.

```python
cat_cols = df.select_dtypes(include='category').columns.drop('tgt')

plt.figure(figsize=(14, 12))
for i, col in enumerate(cat_cols[:9], 1):  # first 9 for clarity
    plt.subplot(3, 3, i)
    sns.countplot(x=col, hue='tgt', data=df, palette='Set2')
    plt.title(f"{col} vs Target")
    plt.xticks(rotation=45)
    plt.xlabel(col)
    plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

These plots compare the category frequencies for **reoffenders (`y`)** and **non-reoffenders (`n`)** across major categorical features.

#### A-group (`a2`, `a3`, `a4`)
- **`a2`** â†’ The category **`x`** dominates and has **more both reoffenders and non-reoffenders (`n`)**,  
  while categories such as `c` and `p` show less reoffending (`y`).  
  Overall, the variable is **imbalanced**, but some categories show clear behavioural differences.
- **`a3`** â†’ Several categories (`s`, `t`, and `y`) have noticeably **higher `y` counts**,  
  this feature for `t` has a **moderate relationship** with the target.
- **`a4`** â†’ Category **`n`** is clearly linked to **non-reoffenders**,  
  while other categories show mixed results. This pattern indicates a **potential protective effect** for `a4 = n`.

#### B Feature
- Category **`f`** has many more **reoffenders** compared to **`t`**.

#### C-group (`c1`, `c2`, `c3`)
- **`c1`** â†’ Categories `a` and `d` show more reoffending (`y`),  
  whereas `e` `f` and `p` are linked to lower reoffending.  
  Moderate variation between groups.  
- **`c2`** â†’ Category **`c`** has a much higher share of **reoffenders**,  
  whereas `d` and `f` are more balanced.  
  This makes `c2` a **strong candidate for predictive modelling**.  
- **`c3`** â†’ Categories `n`, `y` and `w` are clearly linked with **higher reoffending**,  
  while others (e.g. `g`, `r`, `u`) show more balanced outcomes.

#### D-group (`d3`, `d4`)
- **`d3`** â†’ Categories `c`,`f` and `r` have only reoffenders,  
  while others have also non-reoffending instances.  
- **`d4`** â†’ Shows visible variation; categories `i` and `y` have more reoffenders,  
  meanwhile `f`, `g`, `h` have only reoffenders, suggesting this feature might also contribute to prediction.

---

Overall, categorical variablesâ€”particularly `a2`, `a4`, and `d3` carry valuable information 
for distinguishing between reoffenders and non-reoffenders.

### Numeric Feature Exploration (`a1`, `d1`, `d2`)

The histograms above display the frequency distributions for the three numeric features.

- **a1:** Highly right-skewed with several extreme values above 30. Most observations fall between 0 and 10, indicating a long positive tail.  
- **d1:** Slightly right-skewed but more concentrated. The majority of values lie between 4 and 8, forming a clear peak around 5.

```python
categorical_cols = [c for c in df.select_dtypes('category').columns if c != 'tgt']

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index, palette='pastel')
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# target distribution
tgt_counts = df['tgt'].value_counts()
tgt_pct = (tgt_counts / len(df) * 100).round(2)
tgt_summary = pd.DataFrame({'Count': tgt_counts, 'Percentage (%)': tgt_pct})
display(tgt_summary)

# visualise
plt.figure(figsize=(5,4))
sns.countplot(x='tgt', hue='tgt', data=df, palette='Set2', legend=False)
plt.title("Reoffending Distribution (Target)")
plt.xlabel("Reoffended?")
plt.ylabel("Count")
plt.show()
```
